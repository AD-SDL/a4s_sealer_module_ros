```python
import serial
```

```python
ser = 0

def serial_connection():
    global ser
    ser = serial.Serial()
    ser.baudrate = 19200 
    ser.port = '/dev/ttyUSB0'


    #check to see if port is open or closed
    if (ser.isOpen() == False):
        print ('The Port %s is Open '%ser.portstr)
          #timeout in seconds
        ser.timeout = 10
        ser.open()
    else:
        print ('The Port is closed')

#call the serial_connection() function
serial_connection()
```
